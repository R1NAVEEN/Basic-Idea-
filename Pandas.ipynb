# Data Cleaning in pandas

```python
import pandas as pd
data = pd.read_excel(r"C:\Users\krish\Downloads\Customer Call List.xlsx")
data
```

```python
# First drop the duplicates
data = data.drop_duplicates()
data
```

```python
data = data.drop(columns = 'Not_Useful_Column')
data
```

```python
data['Last_Name'] = data['Last_Name'].str.lstrip('...')
data['Last_Name'] = data['Last_Name'].str.lstrip('/')
data['Last_Name'] = data['Last_Name'].str.rstrip('_')
data

# or we can aslo do like below
# df["Last_Name"].str.strip("123._/")
```

```python
data
```

```python
# Before working on phone number convert all the data in numbers
data['Phone_Number'] = data['Phone_Number'].apply(lambda x: str(x))
data
```

```python
# Assuming your DataFrame is named 'data'
data['Phone_Number'] = data['Phone_Number'].str.replace('[^0-9]', '', regex=True)
data['Phone_Number'] = data['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) 
data
```

```python
data['Phone_Number'] = data['Phone_Number'].str.replace('--', '')
data
```

```python
# Let's look at the address column
data[["Street_Address", "State", "Zip_Code"]] = data["Address"].str.split(',', n=2, expand=True)
data
```

```python
#data = data.drop(columns = "Address")
#data
```

```python
data["Paying Customer"] = data["Paying Customer"].str.replace("Yes", "Y")
data["Paying Customer"] = data["Paying Customer"].str.replace("No", "N")
data
```

```python
data["Do_Not_Contact"] = data["Do_Not_Contact"].str.replace("Yes", "Y")
data["Do_Not_Contact"] = data["Do_Not_Contact"].str.replace("No", "N")
data
```

```python
data = data.replace('N/a', ',')
data
```

```python
# due to nan is a string we cannot replace it 
data = data.replace('NaN', ',')
data
```

```python
# instead fii it
data = data.fillna('')
data
```

```python
data = data.replace(',', '')
data
```

```python
for x in data.index:
    if data.loc[x, "Do_Not_Contact"] == "Y":
        data.drop(x, inplace=True)
data

# we can aslo do
# data.drop(subset = "Do_Not_Contact"), inplace=true)
```

```python
for x in data.index:
    if data.loc[x, "Phone_Number"] == "":
        data.drop(x, inplace=True)
data
```

```python
data.reset_index(drop=True)
```
